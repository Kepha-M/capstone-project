### Purpose of this notebook
This notebook trains and tunes the classification models using the processed data from
`02_preprocessing.ipynb`.

Goals for this stage:
- **Baseline model:** HOG + colour-histogram features fed into a Logistic Regression classifier,
  establishing a non-deep-learning reference point.
- **Primary model:** Fine-tune an **EfficientNetB0** backbone (pretrained on ImageNet) on the malaria
  dataset via transfer learning.
- **Threshold tuning:** Adjust the decision threshold post-training to prioritise recall on the
  parasitized class, since missing an infection carries the highest human cost in this clinical
  triage context.
- **Explainability:** Generate Grad-CAM overlays to highlight the cell regions driving each prediction.

**Target variable:** `cell_label` (1 = parasitized, 0 = uninfected)

Trained model artefacts from this notebook are carried forward into `04_evaluation.ipynb` for
performance assessment and disaggregated analysis.
