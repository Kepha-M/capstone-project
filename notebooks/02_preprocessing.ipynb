### Purpose of this notebook
This notebook prepares the NIH/NLM Malaria Cell Images Dataset for modelling, following the EDA
findings from `01_eda.ipynb`.

Goals for this stage:
- Resize all images to a uniform **224×224 px** input size.
- Split the dataset into train / validation / test sets.
- Apply data augmentation (random rotation, horizontal/vertical flip, colour jitter) to improve
  generalisation across staining variations.
- Extract baseline features (HOG + colour histograms) for the non-deep-learning reference model, and
  prepare image tensors/generators for the EfficientNetB0 transfer-learning model.

**Target variable:** `cell_label` (1 = parasitized, 0 = uninfected)

Outputs of this notebook (processed arrays, generators, and/or serialized splits) feed directly into
`03_modelling.ipynb`.
